# Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("world-happiness-report.csv")
df = df[df['Country name'] == 'United States']
```

```python
df.head()
```

# Correlation Matrix

```python
# Create correlation heat map
sns.heatmap(df.corr().round(2), cmap="Wistia", annot=True)
```

# Happiness

### Happiness vs. time

```python
# Plot graph of happiness over time
ax = sns.regplot(x="year", y="Life Ladder", data=df, line_kws={'color':'red'})
ax.set_title("US happiness rating, 2006-2020")
```

# Log GDP per capita

### Log GDP per capita vs. time

```python
# Plot graph of Log GDP per capita over time
ax = sns.regplot(x="year", y="Log GDP per capita", data=df, line_kws={'color':'red'})
ax.set_title("US log GDP per capita, 2006-2020")
```

### Happiness vs. Log GDP per capita

```python
# Plot graph of Happiness rating versus Log GDP per capita
ax = sns.regplot(x="Log GDP per capita", y="Life Ladder", data=df, line_kws={'color':'red'})
ax.set_title("US happiness rating vs log GDP per capita, 2006-2020")
```

# Summary

Over time, the things that make the US a happier country (social support, freedom, generosity) have been decreasing

Meanwhile, the things that make the US a less happier country (higher log GDP/capita, perceptions of corruption) have been increasing
